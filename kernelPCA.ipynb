### Support vector machines
mainly used for classification, learns how to seprate different groups by forming decision boundaries.
reference: https://www.kaggle.com/code/prashant111/svm-classifier-tutorial/notebook

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

```

```python
X, y = make_moons(n_samples=500, noise = 0.02, random_state=471)

plt.scatter(X[:,0],X[:,1], c=y)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)

plt.title("PCA")
plt.scatter(X_pca[:,0],X_pca[:, 1], c=y)
```

```python
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(kernel = 'rbf', gamma = 1) 
X_kpca = kpca.fit_transform(X)
# what's the gamma mean, and the kernel mean
plt.title("Kernel PCA")
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y)
plt.show()
```
